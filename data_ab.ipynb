```python
def replace_outliers_zscore(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    z_scores = (df[column] - mean) / std
    outliers = np.abs(z_scores) > threshold
    median = df[column].median()
    df.loc[outliers, column] = median
    return df
```

```python
columns_to_check = ["消耗", "展现量", "点击量", "转化数", "拍下订单量", 
                    "拍下订单金额", "成交订单量", "成交订单金额", "收藏店铺量", 
                    "收藏宝贝量", "添加购物车量", "投资回报率"]
```

```python
for column in columns_to_check:
    data = replace_outliers_zscore(data, column)
```

```python
output_file_path = 'data_ad.csv'
data.to_csv(output_file_path, index=False)
```

```python
sns.set(style="whitegrid")
```

```python
def plot_time_series(df, column, title):
    plt.figure(figsize=(14, 7))
    sns.lineplot(x='Date', y=column, data=df, marker='o')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
columns_to_plot = ["Ad_Cost", "Ad_Impressions", "Ad_Clicks", "User_Conversions", "Orders_Placed", 
                   "Order_Amount_Placed", "Orders_Completed", "Order_Amount_Completed", "Store_Favorites", 
                   "Product_Favorites", "Add_To_Cart", "ROI"]
```

```python
for column in columns_to_plot:
    plot_time_series(data_last, column, f'{column} Time Series')
```
